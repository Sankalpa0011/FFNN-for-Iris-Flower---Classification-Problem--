## Import Libraries

```python
import pandas as pd

# Load the dataset
dataset = pd.read_csv("./datasets/iris.csv").values
print(dataset,dataset.shape)
```

```python
data = dataset[:,0:4]
target = dataset[:,4]

print(data.shape, target.shape)
```

```python
print(data)
print(target)
```

## Catergorical Encoding

```python
from keras.utils import to_categorical

new_target = to_categorical(target)
```

```python
print(target[:10])
print(new_target[:10])
```

## Train Test Split

```python
x = data 
y = new_target
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
```

## Model Train

### Sequential Method

#### input_dim = used for 1 dimentional data
#### input_shape = used for 2 dimentional data ( images )

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()                           # empty sequential type neaural network (sequential means, add layers one by one)
#one by one layers add krnn puluwan nisa sequential

                                                      # Input layer kyl ekk aththma na. fist hidden layer ekt kynwa input 4 kyla
model.add(Dense(128, input_dim=4, activation="relu")) # 1st Hidden Layer   # Dense =  Dense layers used as fully connected layer
model.add(Dense(32, activation="relu"))               # 2nd Hidden Layer
model.add(Dense(3, activation="softmax"))             # Output Layer


# Compile
model.compile(loss="categorical_crossentropy", optimizer="adaGrad", metrics=["accuracy"])
# loss ekt catogorical crossentropy ganne ape problem eka classfivication nisa
# matrics accuracy ek aniwarya na. loss ek calculate krgnna helpful wnw.
```

```python
model.summary()
```

```python
for layer in model.layers:
    print("=========================", layer.name, "=========================")
    parameters = layer.get_weights()
    print(parameters)
```

```python
# Train the model
model.fit(x_train, y_train, epochs=5000, validation_split=0.1, batch_size=1)  # default batch size is taking as 32 by keras, mama wenas kala batch size eka 1k widihata. ethkota okkoma data records train wenwa

#validation split ekn wnne  train krnna dapuwagen 1% tygnnwa eka epoch ekk train unt passe uththtra harid kyla.
#ita passe loss ekai accuray ekai blnwa.
```

```python
import matplotlib.pyplot as plt

plt.plot(model.history.history["loss"], label="loss")
plt.plot(model.history.history["val_loss"], label="val loss")
plt.legend()
plt.xlabel("Epoch")
plt.ylabel("loss")
plt.show()
```

```python
plt.plot(model.history.history["accuracy"], label="accuracy")
plt.plot(model.history.history["val_accuracy"], label="val accuracy")
plt.legend()
plt.xlabel("Epoch")
plt.ylabel("loss")
plt.show()
```

## Model Evaluation

```python
import numpy as np

# Predicted target
predicted_target = model.predict(x_test)
print(predicted_target)
print(np.argmax(predicted_target, axis=1)) # 2D array ekaka raw vise maxximum value eka tiyena column number ek gnn puluwan( label eka ganna puluwan )
```

```python
# Actual target
print(y_test)
print(np.argmax(y_test, axis=1))
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(np.argmax(y_test, axis=1), np.argmax(predicted_target, axis=1))
print("Accuracy: ", accuracy)
```

```python
model.evaluate(x_test, y_test)  # evaluate magin loss ekai accuracy ekai ganna puluwan
```
